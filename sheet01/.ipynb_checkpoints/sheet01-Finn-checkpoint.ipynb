```python
# TODO: implement PCA (fill in the blanks in the function below)

# DONE

def pca(data, n_components=None, printEig=False):
    """
    Principal Component Analysis on a p x N data matrix.
    
    Parameters
    ----------
    data : np.ndarray
        Data matrix of shape (p, N).
    n_components : int, optional
        Number of requested components. By default returns all components.
    printEig: bool, if true, prints extra info about eigenvalues
        
    Returns
    -------
    np.ndarray, np.ndarray
        the pca components (shape (n_components, p)) and the projection (shape (n_components, N))

    """
    p = data.shape[0]
    n = data.shape[1]
    
    # set n_components to p by default
    n_components = p if n_components is None else n_components
    assert n_components <= data.shape[0], f"Got n_components larger than dimensionality of data!"
    
    # center the data
    mu = data.mean(axis=1)
    data = data - mu.reshape(p,1)
    
    # compute X times X transpose
    cov = np.matmul(data, data.transpose())
    
    # compute the eigenvectors and eigenvalues
    eigVal, eigVec = np.linalg.eig(cov)
    
    # sort the eigenvectors by eigenvalue and take the n_components largest ones
    if printEig:
        print(f"eigenvalues and sorted indices of {n_components} largest eigenvalues")
        print(eigVal)
    sort = np.argsort(eigVal)[-1:-n_components-1:-1] # get n_components sorted indices in descending order
    if printEig:
        print(sort)
    wTLambda = eigVal[sort]
    wT = eigVec[:,sort]
    components = wT.transpose()
    
    # compute X_projected, the projection of the data to the components
    X_transformed = np.matmul(components,data)
    
    return components, X_transformed  # return the n_components first components and the pca projection of the data
```

### b

```python
# TODO: Compute the first principal component in a robust way: Use the Tukey potential with the scale
# parameter s = 1.5 on the distances. Parameterize the line with the angle  to the x-axis. Plot the error
# as a function of . Interpret the two minima.

def TukeyDist(x,y, s):
    """
    x and y: two data points of which the Tukey potential is returned
    s: parameter for Tukey potential
    return d: Tukey potential of x and y with parameter s
    """
    d = np.linalg.norm(x-y)
    pref = s*s/6.
    if d>s:
        return pref
    else:
        return pref*(1-(1-d*d/(s*s))**3)

def calc_loss(data, s, phi, retLPoints=False, centering=True):
    """
    data: (p,n) with p dimension of features and n number of data points
    s: parameter for Tukey potential
    phi: angle of 2D line to which the data is to be projected
    if centering, centers the data
    return:
        data_z: projected centered component in direction w
            (n_components,n) onto the line with angle phi
        data_proj: actual projected data on w
        l: loss
        if retLPoints: returns list of pointwise losses
    """
    # since the Tukey potential is monotonous, the best projection
    # onto the PCs are still the closest points wT . x
    p = data.shape[0]
    n = data.shape[1]
    
    # centering data
    if centering:
        mu = data.mean(axis=1)
        data = data - mu.reshape(p,1)
    
    # projecting data onto PC(phi)
    wT = np.array([[np.cos(phi),np.sin(phi)]])
    data_z = np.matmul(wT,data) # data in PC coordinates
    data_proj = np.matmul(wT.transpose(),data_z) # projected data
    
    # calculating loss for each pair of points
    data_projT = data_proj.transpose()
    dataT = data.transpose()
    loss_points = np.array([TukeyDist(dataT[i],data_projT[i],s) for i in range(len(dataT))])
    
    loss = loss_points.sum() #loss
    
    if retLPoints:
        return data_z, data_proj, loss_points
    
    return data_z, data_proj, loss
```

```python
phi_list = np.linspace(0,2*np.pi,3*360 + 1)
phi_N = len(phi_list)
s_list = [0.5,1,1.5,2,10]
loss = {}
for s in s_list:
    loss[s] = np.zeros(phi_N)
    for i, phi in enumerate(phi_list):
        loss[s][i] = calc_loss(data, s, phi)[2]
```

```python
plt.title("losses for different s parameters")
plt.xlabel(r"$\phi\,/\,\pi$")
plt.ylabel("loss")
for s in s_list:
    plt.plot(phi_list/np.pi,loss[s],label=f"s = {s}")
plt.legend()
plt.show()
```

Since the lines with angles $\varphi$ and $\varphi + \pi$ are the same, there are two minima $\pi$ apart from each other.

Furthermore, if $s$ is too small, then every point has the same error, unless the PC lies basically on the point. On the other hand, if s is too large,than the potential is not robust anymore. This can be seen in the dent getting closer and closer to the two minima for larger $s$. And for $s=10$ the dent overtakes the two minima.

```python
# optimal value of phi
phi_opt = {}
print("s\t my phi\t\t\t Pablos definition of phi")
for s in s_list:
    phi_opt[s] = phi_list[loss[s].argmin()]
    print(s, '\t', phi_opt[s]/np.pi, '\t',0.5 - phi_opt[s]/np.pi)


# plotting this PC
plt.ylim(-3,15)
plt.title(r"2D data with outliers and robust PC")
plt.xlabel("first feature")
plt.ylabel("second feature")
plt.scatter(data[0],data[1],label="data")
plt.plot(xPC1,yPC1,label=r"$1^{st}$ PC with $L_2$ norm",
        linewidth=5,alpha=0.5)
for s in s_list:
    # points for robust PC
    yPCR = xPC1*np.tan(phi_opt[s])
    plt.plot(xPC1,yPCR,label=rf"robust PC for s={s}")
plt.legend()
plt.show()
```

The difference to Pablos results come from 1. the differently defined angle which leads to different points where the loss is calculated, 2. the centering of the data, and 3. the slighltly different spacing (1000 vs. 1081 points for $\varphi$). When I take the same definition for $\varphi$, take a 1000 $\varphi$ values, and center Pablos data, both calculations give the exact same values.

## 3 RANSAC

$p\in (0,1)$ is the probability of sampling an inlying point. Since one samples with replacement, the probability stays $p$ after drawing a point.

Therefore, the probability of $m$ points beeing inliers is $p^m$. The probability to get at least one outlier-free subset after sampling a subset $r$ times is therefore $1-(1-p^m)^r \overset{!}{>}p_{crit}=0.99$. This is supposed to be larger than some critical probability $p_{crit}$.

Therefore, if $r > \frac{ln(1 - p_{crit})}{ln(1-p^m)} = \frac{ln(0.01)}{ln(1-p^m)}$, the probability of obtaining at least one outlier-free subset is larger than $p_{crit} = 99\%$.
